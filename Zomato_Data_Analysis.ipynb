### **`Join Two dataset using left join`**

```python
final_df = pd.merge (df, df_country, on = 'Country Code', how='left')
```

## **`Plot pie chart containing country names and values `**

## **`Now observe Aggregate rating on the basis of rating color and rating text`**

```python
ratings =final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

Observations

---



1.   Ratings in between 4.5 to 4.9 ---> Excellent
2.   Ratings in between 4.0 to 4.4 ----> Very Good
3.   Ratings in between 3.5 to 3.9 ----> Good
4.   Ratings in between 2.5 to 3.4 ----> Average
5.   Ratings in bbetween 1.8 to 2.4 ----> Poor
6.   Rating 0.0 ----> Not Rated





## **`Plot Bar Chart using Seaborn`**

```python
plt.figure(figsize=(10, 6))
# for setiing background color
sns.set(rc={"axes.facecolor": "#f0f0f0", "figure.facecolor": "#f0f0f0"})

sns.barplot(x="Aggregate rating",
            y="Rating Count",
            hue = "Rating color",
            data = ratings)
```

```python
plt.figure(figsize=(14, 4))


sns.barplot(x="Aggregate rating",
            y="Rating Count",
            hue = "Rating color",
            data = ratings,
            palette = ['white', 'red', 'orange', 'yellow','green', 'green'])
```

Observations

---

1. Not Rated count is very high
2. Maximum number of rating between 2.5 to 4.2






```python
# countplot : plot color wise frequency

sns.countplot(x='Rating color',
              data = ratings,
              palette = ['white', 'red', 'orange', 'yellow','green', 'green']
             )
```
